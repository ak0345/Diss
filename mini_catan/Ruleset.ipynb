**Mini-Catan Ruleset**

  Winning Conditions
  - 5 VP (WIN)
  - Settlements = 1 VP
  - Longest Road = 1 VP 
  - Tie Breakers (In case no more settlement and road spots)
  - - Player with more resources wins
  - - Player with Longest Road wins 
  Inventory
  - 5 settlements max per player (no cities)
  - 10 roads per max player
  - 1 card for each resource adjacent to the second settlement at the start of the game

  Basic Game Rules
  - Place 2 settlements and 2 roads in start
  - All settlement must have 2 roads (edges) of space between them
  - Use 1 six-sided die for rolls
  - Road : 1 Wood + 1 Brick
  - Settlement : 1 Wood + 1 Brick + 1 Wheat + 1 Sheep
  - Discard lower bound half of resource cards if 6 is rolled

  Resources
  - Wood
  - Brick
  - Wheat
  - Sheep
  - Desert (no resources) + Robber

  Robber (roll a 6)
  - Initially at Desert
  - Robber Moves using another dice roll (where 6 means back to desert)
  - Said Hex stops producing resources for the settlement(s) adjacent to it
  - Discard upper bound half of resource cards of all players

  Turn Order
   - Resource Phase : Roll the die for resource collection (only settlements adjacent to the hex with the rolled number receive resources)
   - Trade Phase : May trade with other player or Bank at 2:1 Ratio
   - Building Phase : May spend resources to build settlement or road
  
  Longest Road 
   - Awarded to the player with the longest continuous road (minimum 4 road segments)
   - Provides 1 VP
   - Can switch between players if a longer road is built


**State Space**

 - Resources for each player
 - Victory Points of each player
 - Settlement Placement
 - Road Placement
 - Longest Road
 - Tile Numbers
 - Tile Biome
 - Robber Location


**Action Space**

 - Build Road
 - Build Settlement
 - Trade with Player
 -  - Offer
 -  - Request
 -  - Other Player Response
 -  - Response to Counter Offer
 - Trade with Bank
 -  - Offer
 -  - Request
 - End Turn
